```
from PIL import Image
import numpy as np
```

```
img = Image.open('gdrive/My Drive/cameraman.tif')
img
```

```
img_arr = np.asarray(img, dtype='int32')
img_arr
```

```
img_arr.shape
```

```
a = 63
b = 127
c = 191
d = 255
l = 1
m = 1.3
n = 1.5
o = 1.8
v = l * a
w = m * b
x = n * c
y = o * d
```

```
for i in range(len(img_arr)):
  for j in range(len(img_arr[i])):
    if img_arr[i][j] < a:
      img_arr[i][j] *= l
    elif img_arr[i][j] < b:
      img_arr[i][j] = m * (img_arr[i][j] - a) + v
    elif img_arr[i][j] < c:
      img_arr[i][j] = n * (img_arr[i][j] - b) + w
    elif img_arr[i][j] < d:
      img_arr[i][j] = o * (img_arr[i][j] - c) + x
```

```
new_img = Image.fromarray(np.asarray(img_arr, dtype="uint8"), "L")
```
