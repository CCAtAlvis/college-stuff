```
from PIL import Image
import numpy as np
```

```
img = Image.open('gdrive/My Drive/cameraman.tif')
img
```

```
img_arr = np.asarray(img, dtype="int32")
img_arr
```

```
img_arr.shape
```

```
def bits(img_arr):
  bit_arr = []
  for i in range(len(img_arr)):
    for j in range(len(img_arr[i])):
      bit_arr.append(np.binary_repr(img_arr[i][j], width=8))
  return bit_arr
```

```
bits_arr = bits(img_arr)
print(bits_arr)
```

```
def img_plane(bits_arr, n):
  new_bits_arr = []
  for i in range(len(bits_arr)):
    if bits_arr[i][n] == '1':
      new_bits_arr.append(255)
    else:
      new_bits_arr.append(0)
  new_bits_arr = np.reshape(new_bits_arr, (256, 256))
  return new_bits_arr
```

```
new_bits_arr1 = img_plane(bits_arr, 4)
new_img = Image.fromarray(np.asarray(new_bits_arr1, dtype="uint8"), "L")
new_img
```

```
new_bits_arr = img_plane(bits_arr, 7)
new_img1 = Image.fromarray(np.asarray(new_bits_arr, dtype="uint8"), "L")
new_img1
```
